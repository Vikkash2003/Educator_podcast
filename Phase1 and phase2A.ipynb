## ðŸ› ï¸ STEP 5: Create Helper Functions and Utilities

Define utility functions that will be used throughout the project.

````python
# Logging utility function
def log_message(message, level="INFO", save_to_file=True):
    """
    Log a message with timestamp to console and optionally to file.

    Args:
        message (str): Message to log
        level (str): Log level (INFO, WARNING, ERROR, SUCCESS)
        save_to_file (bool): Whether to save to log file
    """
    timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

    # Color codes for different log levels
    colors = {
        'INFO': '\033[94m',      # Blue
        'WARNING': '\033[93m',   # Yellow
        'ERROR': '\033[91m',     # Red
        'SUCCESS': '\033[92m',   # Green
    }
    reset = '\033[0m'

    # Format message
    color = colors.get(level, colors['INFO'])
    formatted_message = f"{color}[{timestamp}] [{level}]{reset} {message}"

    # Print to console
    print(formatted_message)

    # Save to file
    if save_to_file:
        log_file = os.path.join(LOGS_DIR, 'phase1_log.txt')
        with open(log_file, 'a', encoding='utf-8') as f:
            f.write(f"[{timestamp}] [{level}] {message}\n")

print("âœ… Logging function created")
````

````python
# File validation utility
def validate_file_path(file_path, extension=None):
    """
    Validate if a file exists and optionally check its extension.

    Args:
        file_path (str): Path to file
        extension (str): Expected file extension (e.g., '.pdf')

    Returns:
        bool: True if valid, False otherwise
    """
    # Check if file exists
    if not os.path.exists(file_path):
        log_message(f"File not found: {file_path}", level="ERROR")
        return False

    # Check if it's a file (not a directory)
    if not os.path.isfile(file_path):
        log_message(f"Path is not a file: {file_path}", level="ERROR")
        return False

    # Check extension if specified
    if extension:
        if not file_path.lower().endswith(extension.lower()):
            log_message(f"Invalid file extension. Expected {extension}: {file_path}", level="ERROR")
            return False

    return True

print("âœ… File validation function created")
````

````python
# JSON save/load utilities
def save_json(data, file_path, indent=2):
    """
    Save data to JSON file with error handling.

    Args:
        data (dict): Data to save
        file_path (str): Path to save JSON file
        indent (int): Indentation for pretty printing

    Returns:
        bool: True if successful, False otherwise
    """
    try:
        with open(file_path, 'w', encoding='utf-8') as f:
            json.dump(data, f, indent=indent, ensure_ascii=False)
        log_message(f"Data saved to: {file_path}", level="SUCCESS")
        return True
    except Exception as e:
        log_message(f"Error saving JSON: {str(e)}", level="ERROR")
        return False

def load_json(file_path):
    """
    Load data from JSON file with error handling.

    Args:
        file_path (str): Path to JSON file

    Returns:
        dict or None: Loaded data or None if error
    """
    try:
        if not validate_file_path(file_path, extension='.json'):
            return None

        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)
        log_message(f"Data loaded from: {file_path}", level="SUCCESS")
        return data
    except Exception as e:
        log_message(f"Error loading JSON: {str(e)}", level="ERROR")
        return None

print("âœ… JSON utility functions created")
````

````python
# Directory listing utility
def list_files_in_directory(directory, extension=None, recursive=False):
    """
    List all files in a directory with optional filtering.

    Args:
        directory (str): Directory path
        extension (str): Filter by file extension (e.g., '.pdf')
        recursive (bool): Search subdirectories

    Returns:
        list: List of file paths
    """
    files = []

    try:
        if recursive:
            for root, dirs, filenames in os.walk(directory):
                for filename in filenames:
                    if extension is None or filename.lower().endswith(extension.lower()):
                        files.append(os.path.join(root, filename))
        else:
            for filename in os.listdir(directory):
                file_path = os.path.join(directory, filename)
                if os.path.isfile(file_path):
                    if extension is None or filename.lower().endswith(extension.lower()):
                        files.append(file_path)

        return sorted(files)

    except Exception as e:
        log_message(f"Error listing files: {str(e)}", level="ERROR")
        return []

print("âœ… Directory listing function created")
````

````python
# Format file size utility
def format_file_size(size_bytes):
    """
    Convert bytes to human-readable format.

    Args:
        size_bytes (int): Size in bytes

    Returns:
        str: Formatted size string
    """
    for unit in ['B', 'KB', 'MB', 'GB', 'TB']:
        if size_bytes < 1024.0:
            return f"{size_bytes:.2f} {unit}"
        size_bytes /= 1024.0
    return f"{size_bytes:.2f} PB"

print("âœ… File size formatting function created")
````

````python
# Progress tracking utility
class ProgressTracker:
    """
    Track progress of multi-step operations.
    """
    def __init__(self, total_steps, description="Processing"):
        self.total_steps = total_steps
        self.current_step = 0
        self.description = description
        self.start_time = datetime.now()

    def update(self, step_name=""):
        """Update progress by one step."""
        self.current_step += 1
        percentage = (self.current_step / self.total_steps) * 100
        elapsed = (datetime.now() - self.start_time).total_seconds()

        if self.current_step < self.total_steps:
            eta = (elapsed / self.current_step) * (self.total_steps - self.current_step)
            log_message(
                f"{self.description}: {self.current_step}/{self.total_steps} "
                f"({percentage:.1f}%) - {step_name} - ETA: {eta:.1f}s",
                level="INFO"
            )
        else:
            log_message(
                f"{self.description}: Complete! ({elapsed:.1f}s total)",
                level="SUCCESS"
            )

print("âœ… Progress tracker class created")
````

## ðŸ“¥ STEP 2: Download Utility Functions

Create helper functions for downloading and validating PDFs.

````python
def download_pdf(url, save_path, timeout=60):
    """
    Download a PDF from URL and save to specified path.

    Args:
        url (str): URL of the PDF
        save_path (str): Path to save the downloaded PDF
        timeout (int): Download timeout in seconds

    Returns:
        dict: Download status and metadata
    """
    result = {
        'success': False,
        'url': url,
        'save_path': save_path,
        'file_size': 0,
        'error': None
    }

    try:
        print(f"ðŸ“¥ Downloading: {os.path.basename(save_path)}")
        print(f"   URL: {url}")

        # Download with progress
        response = requests.get(url, timeout=timeout, stream=True)
        response.raise_for_status()

        # Get file size
        total_size = int(response.headers.get('content-length', 0))

        # Save file
        with open(save_path, 'wb') as f:
            if total_size == 0:
                f.write(response.content)
            else:
                downloaded = 0
                for chunk in response.iter_content(chunk_size=8192):
                    if chunk:
                        f.write(chunk)
                        downloaded += len(chunk)
                        # Show progress
                        percent = (downloaded / total_size) * 100
                        if percent % 25 < 1:  # Print every 25%
                            print(f"   Progress: {percent:.0f}%")

        # Get actual file size
        result['file_size'] = os.path.getsize(save_path)
        result['success'] = True
        print(f"   âœ… Downloaded successfully ({result['file_size'] / 1024 / 1024:.2f} MB)")

    except requests.exceptions.Timeout:
        result['error'] = 'Download timeout'
        print(f"   âŒ Error: Download timeout")
    except requests.exceptions.RequestException as e:
        result['error'] = str(e)
        print(f"   âŒ Error: {str(e)}")
    except Exception as e:
        result['error'] = str(e)
        print(f"   âŒ Unexpected error: {str(e)}")

    return result

print("âœ… Download function created")
````

````python
def validate_pdf(pdf_path):
    """
    Validate that PDF is text-extractable (not scanned image).

    Args:
        pdf_path (str): Path to PDF file

    Returns:
        dict: Validation results
    """
    validation = {
        'is_valid': False,
        'total_pages': 0,
        'text_extractable': False,
        'sample_text_length': 0,
        'file_size_mb': 0,
        'errors': []
    }

    try:
        # Check file exists and is readable
        if not os.path.exists(pdf_path):
            validation['errors'].append('File not found')
            return validation

        validation['file_size_mb'] = os.path.getsize(pdf_path) / (1024 * 1024)

        # Try to open and extract text from first page
        with pdfplumber.open(pdf_path) as pdf:
            validation['total_pages'] = len(pdf.pages)

            if validation['total_pages'] == 0:
                validation['errors'].append('PDF has no pages')
                return validation

            # Extract text from first page
            first_page = pdf.pages[0]
            sample_text = first_page.extract_text()

            if sample_text:
                validation['sample_text_length'] = len(sample_text.strip())
                validation['text_extractable'] = validation['sample_text_length'] > 100
            else:
                validation['errors'].append('No text extracted - might be scanned PDF')

        # Mark as valid if text is extractable
        validation['is_valid'] = validation['text_extractable']

    except Exception as e:
        validation['errors'].append(f"Validation error: {str(e)}")

    return validation

print("âœ… Validation function created")
````

````python
def get_pdf_metadata(pdf_path):
    """
    Extract metadata from PDF file.

    Args:
        pdf_path (str): Path to PDF file

    Returns:
        dict: PDF metadata
    """
    metadata = {
        'title': '',
        'author': '',
        'subject': '',
        'creator': '',
        'producer': '',
        'creation_date': ''
    }

    try:
        with open(pdf_path, 'rb') as f:
            pdf_reader = PyPDF2.PdfReader(f)
            pdf_info = pdf_reader.metadata

            if pdf_info:
                metadata['title'] = pdf_info.get('/Title', '') or ''
                metadata['author'] = pdf_info.get('/Author', '') or ''
                metadata['subject'] = pdf_info.get('/Subject', '') or ''
                metadata['creator'] = pdf_info.get('/Creator', '') or ''
                metadata['producer'] = pdf_info.get('/Producer', '') or ''

                # Handle creation date
                creation_date = pdf_info.get('/CreationDate', '')
                if creation_date:
                    metadata['creation_date'] = str(creation_date)

    except Exception as e:
        print(f"   âš ï¸  Could not extract metadata: {str(e)}")

    return metadata

print("âœ… Metadata extraction function created")
````

````python
def validate_pdf(pdf_path):
    """
    Validate that PDF is text-extractable (not scanned image).

    Args:
        pdf_path (str): Path to PDF file

    Returns:
        dict: Validation results
    """
    validation = {
        'is_valid': False,
        'total_pages': 0,
        'text_extractable': False,
        'sample_text_length': 0,
        'file_size_mb': 0,
        'errors': []
    }

    try:
        # Check file exists and is readable
        if not os.path.exists(pdf_path):
            validation['errors'].append('File not found')
            return validation

        validation['file_size_mb'] = os.path.getsize(pdf_path) / (1024 * 1024)

        # Try to open and extract text from first page
        with pdfplumber.open(pdf_path) as pdf:
            validation['total_pages'] = len(pdf.pages)

            if validation['total_pages'] == 0:
                validation['errors'].append('PDF has no pages')
                return validation

            # Extract text from first page
            first_page = pdf.pages[0]
            sample_text = first_page.extract_text()

            if sample_text:
                validation['sample_text_length'] = len(sample_text.strip())
                validation['text_extractable'] = validation['sample_text_length'] > 100
            else:
                validation['errors'].append('No text extracted - might be scanned PDF')

        # Mark as valid if text is extractable
        validation['is_valid'] = validation['text_extractable']

    except Exception as e:
        validation['errors'].append(f"Validation error: {str(e)}")

    return validation

print("âœ… Validation function created")
````

````python
def get_pdf_metadata(pdf_path):
    """
    Extract metadata from PDF file.

    Args:
        pdf_path (str): Path to PDF file

    Returns:
        dict: PDF metadata
    """
    metadata = {
        'title': '',
        'author': '',
        'subject': '',
        'creator': '',
        'producer': '',
        'creation_date': ''
    }

    try:
        with open(pdf_path, 'rb') as f:
            pdf_reader = PyPDF2.PdfReader(f)
            pdf_info = pdf_reader.metadata

            if pdf_info:
                metadata['title'] = pdf_info.get('/Title', '') or ''
                metadata['author'] = pdf_info.get('/Author', '') or ''
                metadata['subject'] = pdf_info.get('/Subject', '') or ''
                metadata['creator'] = pdf_info.get('/Creator', '') or ''
                metadata['producer'] = pdf_info.get('/Producer', '') or ''

                # Handle creation date
                creation_date = pdf_info.get('/CreationDate', '')
                if creation_date:
                    metadata['creation_date'] = str(creation_date)

    except Exception as e:
        print(f"   âš ï¸  Could not extract metadata: {str(e)}")

    return metadata

print("âœ… Metadata extraction function created")
````

````python
def discover_pdfs(base_dir):
    """
    Recursively discover all PDF files in the raw_pdfs directory.

    Args:
        base_dir (str): Base directory to search

    Returns:
        list: List of dictionaries with PDF information
    """
    pdfs = []

    for root, dirs, files in os.walk(base_dir):
        for file in files:
            if file.lower().endswith('.pdf'):
                pdf_path = os.path.join(root, file)

                # Determine subject from directory name
                subject = os.path.basename(root)

                pdfs.append({
                    'filename': file,
                    'path': pdf_path,
                    'subject': subject,
                    'size_mb': os.path.getsize(pdf_path) / (1024 * 1024)
                })

    return sorted(pdfs, key=lambda x: x['subject'])

# Discover all PDFs
pdf_list = discover_pdfs(RAW_PDFS_DIR)

print("ðŸ” PDF DISCOVERY")
print("="*60)
print(f"Total PDFs found: {len(pdf_list)}")
print()

# Group by subject
from collections import defaultdict
by_subject = defaultdict(list)
for pdf in pdf_list:
    by_subject[pdf['subject']].append(pdf)

for subject, pdfs in sorted(by_subject.items()):
    total_size = sum(p['size_mb'] for p in pdfs)
    print(f"{subject.capitalize()}: {len(pdfs)} PDFs ({total_size:.2f} MB)")
    for pdf in pdfs:
        print(f"  â€¢ {pdf['filename']} ({pdf['size_mb']:.2f} MB)")
    print()

print("="*60)
````

````python
def extract_text_from_pdf(pdf_path):
    """
    Extract text from PDF using pdfplumber.

    Args:
        pdf_path (str): Path to PDF file

    Returns:
        dict: Extraction results with metadata
    """
    result = {
        'success': False,
        'raw_text': '',
        'total_pages': 0,
        'chars_extracted': 0,
        'pages_with_text': 0,
        'extraction_method': 'pdfplumber',
        'errors': []
    }

    try:
        with pdfplumber.open(pdf_path) as pdf:
            result['total_pages'] = len(pdf.pages)
            page_texts = []

            for page_num, page in enumerate(pdf.pages, 1):
                try:
                    page_text = page.extract_text()

                    if page_text and len(page_text.strip()) > 0:
                        page_texts.append(page_text)
                        result['pages_with_text'] += 1
                    else:
                        result['errors'].append(f"Page {page_num}: No text extracted")

                except Exception as e:
                    result['errors'].append(f"Page {page_num}: {str(e)}")

            # Combine all page texts
            result['raw_text'] = '\n\n'.join(page_texts)
            result['chars_extracted'] = len(result['raw_text'])
            result['success'] = result['pages_with_text'] > 0

    except Exception as e:
        result['errors'].append(f"PDF opening error: {str(e)}")

    return result

print("âœ… Text extraction function created")
````

````python
def normalize_unicode(text):
    """
    Normalize Unicode characters to standard form.

    Args:
        text (str): Input text

    Returns:
        str: Normalized text
    """
    # Normalize to NFKD form and encode to ASCII (removes accents)
    # Then decode back to UTF-8 for proper handling
    try:
        text = unicodedata.normalize('NFKD', text)
        # Keep UTF-8 characters, don't force ASCII
        return text
    except Exception as e:
        return text

def remove_extra_whitespace(text):
    """
    Remove extra spaces, newlines, and tabs.

    Args:
        text (str): Input text

    Returns:
        str: Cleaned text
    """
    # Replace multiple spaces with single space
    text = re.sub(r' {2,}', ' ', text)

    # Replace multiple newlines with double newline (preserve paragraphs)
    text = re.sub(r'\n{3,}', '\n\n', text)

    # Remove tabs
    text = text.replace('\t', ' ')

    # Remove page break characters
    text = text.replace('\f', '\n')

    return text.strip()

def remove_artifacts(text):
    """
    Remove common PDF extraction artifacts.

    Args:
        text (str): Input text

    Returns:
        str: Cleaned text
    """
    # Remove standalone page numbers
    text = re.sub(r'^\d+\s*$', '', text, flags=re.MULTILINE)

    # Remove common header patterns
    text = re.sub(r'^Chapter \d+.*$', '', text, flags=re.MULTILINE)
    text = re.sub(r'^Page \d+.*$', '', text, flags=re.MULTILINE)

    # Remove figure/table references that appear alone
    text = re.sub(r'\[Figure \d+\]', '', text)
    text = re.sub(r'\[Table \d+\]', '', text)

    # Remove URLs (optional - keep if needed for citations)
    # text = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', text)

    return text

def fix_hyphenation(text):
    """
    Fix words broken across lines with hyphens.

    Args:
        text (str): Input text

    Returns:
        str: Text with fixed hyphenation
    """
    # Fix words split across lines (e.g., "exam-\nple" -> "example")
    text = re.sub(r'(\w+)-\s*\n\s*(\w+)', r'\1\2', text)

    return text

def clean_text(raw_text):
    """
    Apply all cleaning steps to raw text.

    Args:
        raw_text (str): Raw extracted text

    Returns:
        str: Cleaned text
    """
    if not raw_text:
        return ''

    # Apply cleaning steps in order
    text = normalize_unicode(raw_text)
    text = fix_hyphenation(text)
    text = remove_artifacts(text)
    text = remove_extra_whitespace(text)

    return text

print("âœ… Text cleaning functions created")
````

````python
def calculate_quality_metrics(raw_text, clean_text, total_pages):
    """
    Calculate quality metrics for extracted text.

    Args:
        raw_text (str): Raw extracted text
        clean_text (str): Cleaned text
        total_pages (int): Total pages in PDF

    Returns:
        dict: Quality metrics
    """
    metrics = {
        'raw_char_count': len(raw_text),
        'clean_char_count': len(clean_text),
        'raw_word_count': len(raw_text.split()),
        'clean_word_count': len(clean_text.split()),
        'sentence_count': 0,
        'paragraph_count': 0,
        'avg_chars_per_page': 0,
        'avg_words_per_page': 0,
        'text_density': 0,
        'sentence_completeness': 0,
        'extraction_success_rate': 0,
        'quality_flag': 'red'
    }

    if not clean_text or total_pages == 0:
        return metrics

    try:
        # Sentence and paragraph counts
        sentences = sent_tokenize(clean_text)
        metrics['sentence_count'] = len(sentences)

        paragraphs = [p for p in clean_text.split('\n\n') if len(p.strip()) > 0]
        metrics['paragraph_count'] = len(paragraphs)

        # Calculate averages
        metrics['avg_chars_per_page'] = metrics['clean_char_count'] / total_pages
        metrics['avg_words_per_page'] = metrics['clean_word_count'] / total_pages

        # Text density (words per character)
        if metrics['clean_char_count'] > 0:
            metrics['text_density'] = metrics['clean_word_count'] / metrics['clean_char_count']

        # Sentence completeness (sentences ending with proper punctuation)
        complete_sentences = sum(1 for s in sentences if s.strip()[-1] in '.!?')
        if metrics['sentence_count'] > 0:
            metrics['sentence_completeness'] = complete_sentences / metrics['sentence_count']

        # Extraction success rate (compared to expected ~2000 chars per page)
        expected_chars = total_pages * 2000
        metrics['extraction_success_rate'] = min(metrics['clean_char_count'] / expected_chars, 1.0)

        # Quality flag
        if (metrics['extraction_success_rate'] >= 0.90 and
            0.15 <= metrics['text_density'] <= 0.25 and
            metrics['sentence_completeness'] >= 0.95):
            metrics['quality_flag'] = 'green'
        elif (metrics['extraction_success_rate'] >= 0.80 and
              metrics['text_density'] >= 0.10):
            metrics['quality_flag'] = 'yellow'
        else:
            metrics['quality_flag'] = 'red'

    except Exception as e:
        print(f"   âš ï¸  Error calculating metrics: {str(e)}")

    return metrics

print("âœ… Quality assessment function created")
````

````python
# Setup and imports
import os
import json
import pandas as pd
import numpy as np
from datetime import datetime
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

# NLP libraries for validation
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
import re

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize audit report
audit_report = {
    'timestamp': datetime.now().isoformat(),
    'audit_type': 'comprehensive_quality_audit',
    'critical_issues': [],
    'warnings': [],
    'recommendations': [],
    'passed_checks': [],
    'statistics': {}
}

def log_issue(severity, category, message, details=""):
    """Log an issue found during audit"""
    issue = {
        'severity': severity,
        'category': category,
        'message': message,
        'details': details,
        'timestamp': datetime.now().isoformat()
    }

    if severity == 'CRITICAL':
        audit_report['critical_issues'].append(issue)
        print(f"âŒ CRITICAL: {category} - {message}")
    elif severity == 'WARNING':
        audit_report['warnings'].append(issue)
        print(f"âš ï¸  WARNING: {category} - {message}")
    elif severity == 'INFO':
        audit_report['recommendations'].append(issue)
        print(f"â„¹ï¸  INFO: {category} - {message}")

    if details:
        print(f"   Details: {details}")

def log_pass(check_name, details=""):
    """Log a passed check"""
    audit_report['passed_checks'].append({
        'check': check_name,
        'details': details
    })
    print(f"âœ… PASS: {check_name}")
    if details:
        print(f"   {details}")

print("ðŸ” PHASE 1 COMPREHENSIVE QUALITY AUDIT")
print("="*80)
print(f"Audit Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"Objective: Ensure Phase 1 is production-ready with zero errors")
print("="*80)
print()
````

````python
print("\n" + "="*80)
print("SECTION 1: FILE INTEGRITY CHECKS")
print("="*80)

# Define paths
BASE_DIR = os.getcwd()
PROJECT_NAME = "ai_podcast_generator"
RAW_PDFS_DIR = f"{BASE_DIR}/{PROJECT_NAME}/data/raw_pdfs"
EXTRACTED_TEXT_DIR = f"{BASE_DIR}/{PROJECT_NAME}/data/extracted_text"
METADATA_DIR = f"{BASE_DIR}/{PROJECT_NAME}/data/metadata"

# CHECK 1.1: All directories exist
print("\nCheck 1.1: Directory Structure")
print("-" * 80)
required_dirs = [RAW_PDFS_DIR, EXTRACTED_TEXT_DIR, METADATA_DIR]
for directory in required_dirs:
    if os.path.exists(directory):
        log_pass(f"Directory exists: {os.path.basename(directory)}")
    else:
        log_issue('CRITICAL', 'Directory', f"Missing directory: {directory}")

# CHECK 1.2: PDF files integrity
print("\nCheck 1.2: PDF Files Integrity")
print("-" * 80)
pdf_files = []
for root, dirs, files in os.walk(RAW_PDFS_DIR):
    for file in files:
        if file.lower().endswith('.pdf'):
            pdf_path = os.path.join(root, file)
            pdf_files.append(pdf_path)

            # Check file size
            size_mb = os.path.getsize(pdf_path) / (1024 * 1024)
            if size_mb < 0.1:
                log_issue('WARNING', 'PDF Size', f"{file} is very small ({size_mb:.2f} MB)")
            elif size_mb > 100:
                log_issue('WARNING', 'PDF Size', f"{file} is very large ({size_mb:.2f} MB)")

            # Check file is readable
            try:
                with open(pdf_path, 'rb') as f:
                    header = f.read(5)
                    if header != b'%PDF-':
                        log_issue('CRITICAL', 'PDF Format', f"{file} is not a valid PDF")
            except Exception as e:
                log_issue('CRITICAL', 'PDF Access', f"Cannot read {file}: {str(e)}")

log_pass(f"PDF Files Found: {len(pdf_files)}")
audit_report['statistics']['total_pdfs'] = len(pdf_files)

# CHECK 1.3: JSON files integrity
print("\nCheck 1.3: JSON Files Integrity")
print("-" * 80)
json_files = [f for f in os.listdir(EXTRACTED_TEXT_DIR) if f.endswith('_clean.json')]

for json_file in json_files:
    json_path = os.path.join(EXTRACTED_TEXT_DIR, json_file)

    try:
        with open(json_path, 'r', encoding='utf-8') as f:
            data = json.load(f)

        # Check file size
        size_kb = os.path.getsize(json_path) / 1024
        if size_kb < 1:
            log_issue('CRITICAL', 'JSON Size', f"{json_file} is suspiciously small ({size_kb:.2f} KB)")

    except json.JSONDecodeError as e:
        log_issue('CRITICAL', 'JSON Format', f"{json_file} has invalid JSON: {str(e)}")
    except Exception as e:
        log_issue('CRITICAL', 'JSON Access', f"Cannot read {json_file}: {str(e)}")

log_pass(f"JSON Files Found: {len(json_files)}")
audit_report['statistics']['total_json_files'] = len(json_files)

# CHECK 1.4: PDF-JSON correspondence
print("\nCheck 1.4: PDF-JSON Correspondence")
print("-" * 80)
if len(json_files) == len(pdf_files):
    log_pass(f"PDF-JSON match: {len(pdf_files)} PDFs = {len(json_files)} JSON files")
elif len(json_files) < len(pdf_files):
    missing = len(pdf_files) - len(json_files)
    log_issue('CRITICAL', 'Missing JSON', f"{missing} PDFs have no corresponding JSON files")
else:
    extra = len(json_files) - len(pdf_files)
    log_issue('WARNING', 'Extra JSON', f"{extra} JSON files have no corresponding PDFs")
````

````python
print("\n" + "="*80)
print("SECTION 2: DATA STRUCTURE VALIDATION")
print("="*80)

# CHECK 2.1: Required fields in JSON
print("\nCheck 2.1: Required Fields in JSON Files")
print("-" * 80)

required_fields = {
    'metadata': ['source_file', 'subject', 'total_pages', 'extraction_date'],
    'content': ['clean_text', 'word_count', 'char_count', 'sentence_count'],
    'quality_metrics': ['extraction_success_rate', 'quality_flag', 'text_density']
}

all_data = []
for json_file in json_files:
    json_path = os.path.join(EXTRACTED_TEXT_DIR, json_file)

    try:
        with open(json_path, 'r', encoding='utf-8') as f:
            data = json.load(f)

        all_data.append(data)

        # Check top-level sections
        for section in required_fields.keys():
            if section not in data:
                log_issue('CRITICAL', 'Missing Section',
                         f"{json_file} missing '{section}' section")
            else:
                # Check fields within section
                for field in required_fields[section]:
                    if field not in data[section]:
                        log_issue('CRITICAL', 'Missing Field',
                                 f"{json_file} missing '{field}' in '{section}'")

    except Exception as e:
        log_issue('CRITICAL', 'Data Structure', f"Error reading {json_file}: {str(e)}")

log_pass(f"Structure validation complete for {len(all_data)} files")

# CHECK 2.2: Data types validation
print("\nCheck 2.2: Data Types Validation")
print("-" * 80)

for idx, data in enumerate(all_data):
    json_file = json_files[idx]

    try:
        # Check content fields are correct types
        if not isinstance(data.get('content', {}).get('clean_text', ''), str):
            log_issue('CRITICAL', 'Data Type', f"{json_file}: clean_text is not a string")

        if not isinstance(data.get('content', {}).get('word_count', 0), int):
            log_issue('CRITICAL', 'Data Type', f"{json_file}: word_count is not an integer")

        if not isinstance(data.get('quality_metrics', {}).get('extraction_success_rate', 0), (int, float)):
            log_issue('CRITICAL', 'Data Type', f"{json_file}: success_rate is not numeric")

        if not isinstance(data.get('metadata', {}).get('total_pages', 0), int):
            log_issue('CRITICAL', 'Data Type', f"{json_file}: total_pages is not an integer")

    except Exception as e:
        log_issue('CRITICAL', 'Data Type Check', f"Error in {json_file}: {str(e)}")

log_pass(f"Data type validation complete")
````

````python
print("\n" + "="*80)
print("SECTION 3: TEXT CONTENT QUALITY")
print("="*80)

# CHECK 3.1: Empty or minimal text
print("\nCheck 3.1: Text Content Sufficiency")
print("-" * 80)

for idx, data in enumerate(all_data):
    json_file = json_files[idx]
    clean_text = data.get('content', {}).get('clean_text', '')
    word_count = data.get('content', {}).get('word_count', 0)

    # Check for empty text
    if not clean_text or len(clean_text.strip()) == 0:
        log_issue('CRITICAL', 'Empty Text', f"{json_file} has no text content")
    elif word_count < 100:
        log_issue('CRITICAL', 'Minimal Text',
                 f"{json_file} has only {word_count} words (too little)")
    elif word_count < 500:
        log_issue('WARNING', 'Low Word Count',
                 f"{json_file} has only {word_count} words (consider reviewing)")

log_pass("Text sufficiency check complete")

# CHECK 3.2: Text quality issues
print("\nCheck 3.2: Text Quality Issues")
print("-" * 80)

for idx, data in enumerate(all_data):
    json_file = json_files[idx]
    clean_text = data.get('content', {}).get('clean_text', '')

    if clean_text:
        # Check for excessive special characters (might indicate extraction issues)
        special_char_ratio = len(re.findall(r'[^a-zA-Z0-9\s.,!?;:\'"()-]', clean_text)) / len(clean_text)
        if special_char_ratio > 0.1:
            log_issue('WARNING', 'Special Characters',
                     f"{json_file} has {special_char_ratio*100:.1f}% special characters (might have extraction issues)")

        # Check for repeated characters (OCR errors)
        if re.search(r'(.)\1{5,}', clean_text):
            log_issue('WARNING', 'Repeated Characters',
                     f"{json_file} has sequences of repeated characters (possible OCR error)")

        # Check for suspicious patterns
        if clean_text.count('\n') / len(clean_text) > 0.05:
            log_issue('WARNING', 'Excessive Newlines',
                     f"{json_file} has many newlines (might need better cleaning)")

log_pass("Text quality check complete")

# CHECK 3.3: Word count consistency
print("\nCheck 3.3: Word Count Consistency")
print("-" * 80)

for idx, data in enumerate(all_data):
    json_file = json_files[idx]
    clean_text = data.get('content', {}).get('clean_text', '')
    recorded_word_count = data.get('content', {}).get('word_count', 0)

    if clean_text:
        # Recalculate word count
        actual_word_count = len(clean_text.split())

        # Allow 5% variance
        if abs(actual_word_count - recorded_word_count) / recorded_word_count > 0.05:
            log_issue('WARNING', 'Word Count Mismatch',
                     f"{json_file}: recorded={recorded_word_count}, actual={actual_word_count}")

log_pass("Word count consistency check complete")
````

````python
print("\n" + "="*80)
print("SECTION 4: STATISTICAL VALIDATION")
print("="*80)

# Collect all statistics
stats_data = []
for idx, data in enumerate(all_data):
    stats_data.append({
        'filename': json_files[idx],
        'pages': data.get('metadata', {}).get('total_pages', 0),
        'words': data.get('content', {}).get('word_count', 0),
        'sentences': data.get('content', {}).get('sentence_count', 0),
        'chars': data.get('content', {}).get('char_count', 0),
        'success_rate': data.get('quality_metrics', {}).get('extraction_success_rate', 0),
        'text_density': data.get('quality_metrics', {}).get('text_density', 0),
        'quality_flag': data.get('quality_metrics', {}).get('quality_flag', '')
    })

df_stats = pd.DataFrame(stats_data)

# CHECK 4.1: Outlier detection
print("\nCheck 4.1: Outlier Detection")
print("-" * 80)

# Words per page outliers
df_stats['words_per_page'] = df_stats['words'] / df_stats['pages']
mean_wpp = df_stats['words_per_page'].mean()
std_wpp = df_stats['words_per_page'].std()

for idx, row in df_stats.iterrows():
    wpp = row['words_per_page']
    if abs(wpp - mean_wpp) > 2 * std_wpp:
        log_issue('WARNING', 'Statistical Outlier',
                 f"{row['filename']}: {wpp:.0f} words/page (mean={mean_wpp:.0f}, std={std_wpp:.0f})")

log_pass(f"Outlier detection complete (mean words/page: {mean_wpp:.0f})")

# CHECK 4.2: Quality distribution
print("\nCheck 4.2: Quality Distribution Analysis")
print("-" * 80)

quality_counts = df_stats['quality_flag'].value_counts()
red_count = quality_counts.get('red', 0)
yellow_count = quality_counts.get('yellow', 0)
green_count = quality_counts.get('green', 0)

print(f"Quality Distribution:")
print(f"  ðŸŸ¢ Green: {green_count}")
print(f"  ðŸŸ¡ Yellow: {yellow_count}")
print(f"  ðŸ”´ Red: {red_count}")

if red_count > 0:
    log_issue('CRITICAL', 'Quality Issue', f"{red_count} PDFs have RED quality flags")
    red_files = df_stats[df_stats['quality_flag'] == 'red']['filename'].tolist()
    print(f"   Red files: {red_files}")

if green_count / len(df_stats) < 0.5:
    log_issue('WARNING', 'Quality Distribution',
             f"Only {green_count}/{len(df_stats)} PDFs have green quality (below 50%)")
else:
    log_pass(f"Quality distribution acceptable: {green_count}/{len(df_stats)} green")

# CHECK 4.3: Success rate validation
print("\nCheck 4.3: Extraction Success Rate")
print("-" * 80)

avg_success = df_stats['success_rate'].mean()
min_success = df_stats['success_rate'].min()

print(f"Average success rate: {avg_success*100:.1f}%")
print(f"Minimum success rate: {min_success*100:.1f}%")

if avg_success < 0.7:
    log_issue('CRITICAL', 'Low Success Rate',
             f"Average extraction success rate is {avg_success*100:.1f}% (below 70%)")
elif avg_success < 0.8:
    log_issue('WARNING', 'Moderate Success Rate',
             f"Average extraction success rate is {avg_success*100:.1f}% (below 80%)")
else:
    log_pass(f"Extraction success rate excellent: {avg_success*100:.1f}%")

# Store statistics
audit_report['statistics'].update({
    'avg_words_per_page': float(mean_wpp),
    'avg_success_rate': float(avg_success),
    'total_words': int(df_stats['words'].sum()),
    'total_pages': int(df_stats['pages'].sum()),
    'green_count': int(green_count),
    'yellow_count': int(yellow_count),
    'red_count': int(red_count)
})
````

````python
def preprocess_for_summarization(text):
    """
    Additional cleaning specifically for summarization.
    Removes artifacts that BART/T5 might struggle with.

    Args:
        text (str): Input text

    Returns:
        str: Cleaned text
    """
    # Remove timestamps
    text = re.sub(r'\d{1,2}/\d{1,2}/\d{2,4},?\s*\d{1,2}:\d{2}\s*[AP]M', '', text)

    # Remove URLs
    text = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', text)

    # Remove email addresses
    text = re.sub(r'\S+@\S+', '', text)

    # Remove Project Gutenberg boilerplate (first 1000 chars if contains "Gutenberg")
    if 'Gutenberg' in text[:1000]:
        # Find the end of Gutenberg header (usually after "START OF")
        match = re.search(r'\*\*\* START OF .*? \*\*\*', text)
        if match:
            text = text[match.end():]
        else:
            # Fallback: remove first 1000 chars
            text = text[1000:]

    # Remove standalone page numbers
    text = re.sub(r'\n\s*\d{1,4}\s*\n', '\n', text)

    # Remove excessive whitespace
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'\n\s*\n', '\n\n', text)

    return text.strip()

# Apply preprocessing
print("ðŸ§¹ Preprocessing text for summarization...")
print("="*60)

for doc in documents:
    original_length = len(doc['clean_text'])
    doc['preprocessed_text'] = preprocess_for_summarization(doc['clean_text'])
    new_length = len(doc['preprocessed_text'])

    removed_pct = ((original_length - new_length) / original_length) * 100
    print(f"âœ… {doc['source_file'][:40]}...")
    print(f"   Removed {removed_pct:.1f}% of characters (artifacts/noise)")

print("\nâœ… Preprocessing complete")
````

````python
def extractive_summarization_textrank(text, num_sentences=20):
    """
    TextRank algorithm for extractive summarization.
    Based on PageRank applied to sentences.

    Args:
        text (str): Input text
        num_sentences (int): Number of sentences to extract

    Returns:
        str: Extracted summary
    """
    # Tokenize into sentences
    sentences = sent_tokenize(text)

    if len(sentences) <= num_sentences:
        return text

    # Create TF-IDF matrix
    vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)

    try:
        tfidf_matrix = vectorizer.fit_transform(sentences)

        # Calculate similarity matrix
        similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)

        # Create graph
        nx_graph = nx.from_numpy_array(similarity_matrix)

        # Calculate PageRank scores
        scores = nx.pagerank(nx_graph)

        # Rank sentences
        ranked_sentences = sorted(
            ((scores[i], i, s) for i, s in enumerate(sentences)),
            reverse=True
        )

        # Select top sentences and sort by original order
        selected = sorted(ranked_sentences[:num_sentences], key=lambda x: x[1])
        summary = ' '.join([s[2] for s in selected])

        return summary

    except Exception as e:
        print(f"   âš ï¸  TextRank failed: {str(e)}. Returning first {num_sentences} sentences.")
        return ' '.join(sentences[:num_sentences])

print("ðŸ” EXTRACTIVE SUMMARIZATION (TextRank)")
print("="*60)

for doc in tqdm(documents, desc="Extracting summaries"):
    # Calculate target number of sentences (20-30% of original)
    target_sentences = max(15, int(doc['sentence_count'] * 0.25))

    doc['extractive_summary'] = extractive_summarization_textrank(
        doc['preprocessed_text'],
        num_sentences=target_sentences
    )

    doc['extractive_word_count'] = len(doc['extractive_summary'].split())
    compression_ratio = (doc['extractive_word_count'] / doc['word_count']) * 100

    print(f"\nâœ… {doc['source_file'][:40]}...")
    print(f"   Original: {doc['word_count']:,} words")
    print(f"   Extracted: {doc['extractive_word_count']:,} words ({compression_ratio:.1f}% of original)")

print("\nâœ… Extractive summarization complete")
````

````python
def abstractive_summarization_t5(text, tokenizer, model, device, max_length=150, min_length=50):
    """
    Generate abstractive summary using T5-Small model.

    Args:
        text (str): Input text
        tokenizer: T5 tokenizer
        model: T5 model
        device: 'cuda' or 'cpu'
        max_length (int): Maximum summary length
        min_length (int): Minimum summary length

    Returns:
        str: Abstractive summary
    """
    if tokenizer is None or model is None:
        # Fallback
        words = text.split()[:min_length]
        return ' '.join(words)

    # T5 requires "summarize: " prefix
    input_text = "summarize: " + text

    # Tokenize (T5 handles up to 512 tokens)
    words = text.split()
    chunk_size = 400  # Conservative for T5

    summaries = []

    for i in range(0, len(words), chunk_size):
        chunk = ' '.join(words[i:i + chunk_size])

        if len(chunk.split()) < 30:
            continue

        try:
            # Prepare input
            input_text = "summarize: " + chunk
            inputs = tokenizer.encode(
                input_text,
                return_tensors="pt",
                max_length=512,
                truncation=True
            ).to(device)

            # Generate summary
            outputs = model.generate(
                inputs,
                max_length=max_length,
                min_length=min_length,
                length_penalty=2.0,
                num_beams=4,
                early_stopping=True
            )

            # Decode
            summary = tokenizer.decode(outputs[0], skip_special_tokens=True)
            summaries.append(summary)

            # Clear cache
            if device == "cuda":
                torch.cuda.empty_cache()

        except Exception as e:
            print(f"   âš ï¸ Chunk {i//chunk_size + 1} failed: {str(e)[:50]}")
            continue

    if not summaries:
        # Fallback to first portion
        return ' '.join(text.split()[:min_length * 2])

    # Combine summaries
    combined = ' '.join(summaries)

    # If still too long, summarize again
    if len(combined.split()) > max_length * 3:
        try:
            input_text = "summarize: " + combined
            inputs = tokenizer.encode(
                input_text,
                return_tensors="pt",
                max_length=512,
                truncation=True
            ).to(device)

            outputs = model.generate(
                inputs,
                max_length=max_length * 2,
                min_length=max_length,
                length_penalty=2.0,
                num_beams=4
            )

            return tokenizer.decode(outputs[0], skip_special_tokens=True)
        except:
            return ' '.join(combined.split()[:max_length * 2])

    return combined

print("âœ… T5 summarization function defined")
````

````python
def create_hybrid_summary(doc):
    """
    Create a hybrid summary combining all techniques.

    Structure:
    1. Abstractive summary (overview)
    2. Key extractive sentences (details)
    3. Concept list (reference)

    Args:
        doc (dict): Document with all summaries

    Returns:
        dict: Structured hybrid summary
    """
    hybrid = {
        'title': doc['source_file'].replace('_clean.json', ''),
        'subject': doc['subject'],
        'overview': doc['abstractive_summary'],
        'key_points': [],
        'key_concepts': [c['term'] for c in doc['key_concepts'][:15]],
        'noun_phrases': doc['noun_phrases'][:10],
        'metadata': {
            'original_pages': doc['total_pages'],
            'original_words': doc['word_count'],
            'summary_words': doc['abstractive_word_count'],
            'compression_ratio': (doc['abstractive_word_count'] / doc['word_count']) * 100
        }
    }

    # Extract key sentences from extractive summary
    extractive_sentences = sent_tokenize(doc['extractive_summary'])

    # Select 8-12 most informative sentences
    # Prioritize longer sentences (more information)
    scored_sentences = [
        (len(sent.split()), sent)
        for sent in extractive_sentences
        if len(sent.split()) >= 10  # Meaningful sentences
    ]
    scored_sentences.sort(reverse=True)

    hybrid['key_points'] = [sent for _, sent in scored_sentences[:10]]

    # Calculate total summary length
    total_summary_text = (
        hybrid['overview'] + ' ' +
        ' '.join(hybrid['key_points'])
    )
    hybrid['total_summary_words'] = len(total_summary_text.split())

    return hybrid

print("ðŸ”„ CREATING HYBRID SUMMARIES")
print("="*60)

for doc in documents:
    doc['hybrid_summary'] = create_hybrid_summary(doc)

    print(f"\nâœ… {doc['source_file'][:40]}...")
    print(f"   Overview: {len(doc['hybrid_summary']['overview'].split())} words")
    print(f"   Key points: {len(doc['hybrid_summary']['key_points'])} sentences")
    print(f"   Concepts: {len(doc['hybrid_summary']['key_concepts'])} terms")
    print(f"   Total summary: {doc['hybrid_summary']['total_summary_words']} words")
    print(f"   Compression: {doc['hybrid_summary']['metadata']['compression_ratio']:.1f}%")

print("\nâœ… Hybrid summaries complete")
````

````python
def calculate_rouge_scores(summary, reference, max_reference_length=10000):
    """
    Calculate ROUGE scores for summary quality.

    Args:
        summary (str): Generated summary
        reference (str): Original text (reference)
        max_reference_length (int): Limit reference length for efficiency

    Returns:
        dict: ROUGE scores
    """
    scorer = rouge_scorer.RougeScorer(
        ['rouge1', 'rouge2', 'rougeL'],
        use_stemmer=True
    )

    # Limit reference length for efficiency
    if len(reference.split()) > max_reference_length:
        reference = ' '.join(reference.split()[:max_reference_length])

    scores = scorer.score(reference, summary)

    return {
        'rouge1_f': scores['rouge1'].fmeasure,
        'rouge1_p': scores['rouge1'].precision,
        'rouge1_r': scores['rouge1'].recall,
        'rouge2_f': scores['rouge2'].fmeasure,
        'rougeL_f': scores['rougeL'].fmeasure
    }

print("ðŸ“Š CALCULATING ROUGE SCORES")
print("="*60)
print("Evaluating summary quality...\n")

for doc in tqdm(documents, desc="Calculating ROUGE"):
    # Calculate for abstractive summary
    doc['rouge_scores_abstractive'] = calculate_rouge_scores(
        doc['abstractive_summary'],
        doc['preprocessed_text']
    )

    # Calculate for hybrid summary
    hybrid_text = (
        doc['hybrid_summary']['overview'] + ' ' +
        ' '.join(doc['hybrid_summary']['key_points'])
    )
    doc['rouge_scores_hybrid'] = calculate_rouge_scores(
        hybrid_text,
        doc['preprocessed_text']
    )

    print(f"\nâœ… {doc['source_file'][:40]}...")
    print(f"   Abstractive ROUGE-1: {doc['rouge_scores_abstractive']['rouge1_f']:.3f}")
    print(f"   Hybrid ROUGE-1: {doc['rouge_scores_hybrid']['rouge1_f']:.3f}")

print("\nâœ… ROUGE evaluation complete")
````

````python
#Analyze summaries and select 2-3 best PDFs for podcast creation.
def calculate_selection_score(doc):
    """
    Calculate a score for PDF selection based on multiple criteria.

    Criteria:
    - ROUGE score (quality)
    - Summary length (not too short/long)
    - Quality flag (green > yellow > red)
    - Subject diversity (bonus)

    Args:
        doc (dict): Document data

    Returns:
        float: Selection score (higher is better)
    """
    score = 0

    # ROUGE score (0-50 points)
    rouge_score = doc['rouge_scores_hybrid']['rouge1_f']
    score += rouge_score * 50

    # Summary length appropriateness (0-20 points)
    summary_words = doc['hybrid_summary']['total_summary_words']
    if 800 <= summary_words <= 2000:  # Ideal range
        score += 20
    elif 500 <= summary_words <= 3000:  # Acceptable
        score += 10

    # Quality flag (0-20 points)
    quality_scores = {'green': 20, 'yellow': 10, 'red': 0}
    score += quality_scores.get(doc['quality_flag'], 0)

    # Page count preference (0-10 points)
    # Prefer 15-25 pages (optimal for podcast)
    pages = doc['total_pages']
    if 15 <= pages <= 25:
        score += 10
    elif 10 <= pages <= 30:
        score += 5

    return score

print("ðŸŽ¯ SELECTING BEST PDFs FOR PODCAST GENERATION")
print("="*60)

# Calculate scores
for doc in documents:
    doc['selection_score'] = calculate_selection_score(doc)

# Sort by score
sorted_docs = sorted(documents, key=lambda x: x['selection_score'], reverse=True)

print("\nDocument Rankings:")
print("-" * 60)
for i, doc in enumerate(sorted_docs, 1):
    print(f"{i}. {doc['source_file'][:45]}...")
    print(f"   Score: {doc['selection_score']:.1f} | Subject: {doc['subject']} | "
          f"Quality: {doc['quality_flag']} | ROUGE: {doc['rouge_scores_hybrid']['rouge1_f']:.3f}")
    print(f"   Pages: {doc['total_pages']} | Summary: {doc['hybrid_summary']['total_summary_words']} words")
    print()

# Select top 3, ensuring subject diversity
selected_docs = []
selected_subjects = set()

for doc in sorted_docs:
    # Prefer different subjects
    if doc['subject'] not in selected_subjects or len(selected_docs) < 2:
        selected_docs.append(doc)
        selected_subjects.add(doc['subject'])

    if len(selected_docs) >= 3:
        break

print("\n" + "="*60)
print("ðŸŽ¯ SELECTED PDFs FOR PODCAST GENERATION")
print("="*60)

for i, doc in enumerate(selected_docs, 1):
    print(f"\n{i}. {doc['source_file']}")
    print(f"   Subject: {doc['subject']}")
    print(f"   Quality: {doc['quality_flag']} (ROUGE-1: {doc['rouge_scores_hybrid']['rouge1_f']:.3f})")
    print(f"   Original: {doc['word_count']:,} words ({doc['total_pages']} pages)")
    print(f"   Summary: {doc['hybrid_summary']['total_summary_words']:,} words")
    print(f"   Selection Score: {doc['selection_score']:.1f}/100")

# Save selection report
selection_report = {
    'timestamp': datetime.now().isoformat(),
    'selection_criteria': {
        'rouge_weight': 50,
        'length_weight': 20,
        'quality_weight': 20,
        'pages_weight': 10
    },
    'all_documents': [
        {
            'filename': doc['source_file'],
            'score': doc['selection_score'],
            'rank': i
        }
        for i, doc in enumerate(sorted_docs, 1)
    ],
    'selected_documents': [
        {
            'filename': doc['source_file'],
            'subject': doc['subject'],
            'score': doc['selection_score'],
            'rouge1': doc['rouge_scores_hybrid']['rouge1_f'],
            'summary_words': doc['hybrid_summary']['total_summary_words']
        }
        for doc in selected_docs
    ]
}

selection_path = os.path.join(METADATA_DIR, 'pdf_selection_report.json')
with open(selection_path, 'w') as f:
    json.dump(selection_report, f, indent=2)

print(f"\nðŸ’¾ Selection report saved to: {selection_path}")
````
